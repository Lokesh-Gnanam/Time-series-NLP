```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv('/Users/mohamedsabeer/Desktop/TS & NLP/TSNLP D1 - C2.csv')
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])

df.set_index('DATE', inplace=True)

df = df.sort_index()

decomposition = seasonal_decompose(
    df['Consumption'],
    model='additive',
    period=12
)

decomposition.plot()
plt.tight_layout()
plt.show()
```
